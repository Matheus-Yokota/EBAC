####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Como na hora do plot o eixo x da gestação estava ficando sobreposto tive que fazer umas pesquisas de como arrumar isso.

# Primeiro eu coloquei os itens da GESTACAO em uma lista, removi os dados duplidados e os dados vazios.

valores_gestacao = df_csv['GESTACAO'].drop_duplicates().dropna().tolist()
```

```python
df_csv['APGAR1'].value_counts()
```

```python
valores_gestacao
```

```python
# Primeiro fechar os gráficos abertos
# Depois realizo o plot dos gráficos.

plt.close('all')

plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(nrows=2, ncols=2)

sns.countplot(ax=axes[0, 0], x='SEXO', data=df_csv)
sns.countplot(ax=axes[0, 1], x='GESTACAO', data=df_csv)
sns.countplot(ax=axes[1, 0], x='ESTCIVMAE', data=df_csv)
sns.countplot(ax=axes[1, 1], x='CONSULTAS', data=df_csv)

axes[0, 1].set_xticklabels(valores_gestacao, rotation=330)

plt.subplots_adjust(wspace=0.3, hspace=0.5)
```
