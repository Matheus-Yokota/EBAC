````python
# Importando as bibliotecas

import pandas as pd
import numpy as np
import seaborn as sns

import plotly.express as px
from plotly import graph_objects

from plotly import graph_objects
from plotly.graph_objs import Layout
````

#### 1. Carregue os dados da tarefa anterior

````python
# Carregando a base de dados produzida na tarefa anterior

caminho = '/content/ativos_yahoo.csv'
df = pd.read_csv(caminho, header=[0, 1])
````

````python
# Cópia do dataframe original

df_1 = df.copy()

# Manipulando o dataframe para ficar da forma desejada
df_1 = df.dropna().set_index('Price').stack().reset_index()
df_1.head()
````

````python
# Ao tentar alterar a data para um formato mais amigável verificou-se que esta virou uma tupla
# Para manipula-la transformei em uma string

df_1['Price'] = df_1.Price.map(lambda x: str(x))
````

````python
# Com a string retirei a virgula e os parenteses utilizando um laço de repetição

tamanho = len(df_1.Price)

i = 0

while i < tamanho:
    df_1.Price[i] = df_1.Price[i].replace(',', '')
    df_1.Price[i] = df_1.Price[i].replace('(', '')
    df_1.Price[i] = df_1.Price[i].replace(')', '')
    i += 1
````

````python
# Por fim, transformei no tipo data

df_1['Price'] = pd.to_datetime(df_1['Price'])
df_1.head()
````

````python
# Alterando o nome das colunas Price e level_1

df_1 = df_1.rename(columns = {'Price': 'Date'})
df_1 = df_1.rename(columns = {'level_1': 'Ticker'})

# Agora o dataframe está pronto para uso

df_1.head()
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Criando o Dataframe base das minhas operações

df_2 = df_1.pivot(index='Date', columns='Ticker', values='Adj Close')
df_2.head()
````

````python
# Usando o rolling para obtera a media móvel e o desvio pardrão móvel

df_media = df_2.rolling(90).mean()
df_std = df_2.rolling(90).std()
````

````python
# Plot da media móvel

df_media_1 = df_media.stack().reset_index()
fig = px.line(df_media_1, x='Date', y=0, color='Ticker')
fig.show()
````

````python
# Plot da desvio padrão móvel

df_std_1 = df_std.stack().reset_index()
fig = px.line(df_std_1, x='Date', y=0, color='Ticker')
fig.show()
````

### Insights:

- Média:
    - Observando a média móvel dos ativos escolhidos na atividade anterior verifica-se uma flutuabilidade menor do que analisando o valor diário dos ativos;
    - Analisando os ativos, o valor das ações da Magazine Luisa estiveram abaixo das ações da CEMIG e do Bradesco durante todo período analisado;
    - No período analisado, inicialmente o valor das ações da CEMIG estavam abaixo das ações do Bradesco, contudo, em pouco tempo o valor do Bradesco foi ultrapassado.

- Desvio padrão:
    - Diferente do que é observado para a média, o desvio padrão móvel já apresenta uma flutuabilidade muito maior.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
# Selecionando as ações da CEMIG

df_3 = df_1[df_1['Ticker'] == 'CMIG3.SA']
````

````python
# Obtendo os últimos 60 dias desta ação

df_3 = (df_3.tail(60)
        .reset_index(drop=True))
df_3.head()
````

````python
# Dicionário com parâmetros que serão utilizados

graph = {
    'x': df_3.Date,
    'open': df_3.Open,
    'close': df_3.Close,
    'high': df_3.High,
    'low': df_3.Low,
    'type': 'candlestick',
    'name': 'CMIG3',
    'showlegend': True
}
````

````python
# Plot do gráfico

layout = Layout(paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='rgba(0,0,0,0)')

grafico = graph_objects.Figure(data = [graph],
                               layout_title='CEMIG',
                               layout=layout)

grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
