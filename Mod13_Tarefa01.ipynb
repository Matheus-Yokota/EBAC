````python
# Antes de começar a modelagem de regressão deve-se realizar a limpeza do banco de dados.

# Como a base de dados é bem extensa, eu optei por remover as linhas com dados faltantes

df = df.dropna()

# Removendo linhas duplicadas

df = df.drop_duplicates()

# Resetando o index

df = df.reset_index()
````

````python
df.info()
````

````python
# Como a base de dados é bem grande, vou realizar a modelagem apenas com a segunda metade

# Primeiro irei realizar uma cópia da base de dados.

df_1 = df.copy()

# Segundo, selecionando a segunda metade da base de dados

df_1 = df_1.iloc[6213:,:]
````

````python
# Com a base de dados pronta, inicia-se a modelagem o patsy

# A priori seriam utilizadas todas as variáveis para explicar o modelo, contudo a variável index, unnamed, data_ref e id_cliente, não farão parte da modelagem, pois não conseguem explicar o valor da renda de um indivíduo.

# As variáveis sexo, tipo_renda, educacao, estado_civil e tipo_residencia são variáveis categóricas
# A casela de referencia é a categoria com maior frequencia

y, X = patsy.dmatrices('np.log(renda) ~ C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda, Treatment("Assalariado")) + \
                       C(educacao, Treatment("Secundário")) + C(estado_civil, Treatment("Casado")) + C(tipo_residencia, Treatment("Casa")) + idade + tempo_emprego + qt_pessoas_residencia + 0', data=df_1)
````

````python
# Usando o statsmodel para ver como ficou a regressão

# Aparentemente as variáveis utilizadas para explicar o modelo de regressão fazem sentido.

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

### Insights:

A variável ```tipo_residencia``` não apresenta significância estatística para o modelo, dado que o p-value é maior que 0.05 em todas as categorias.

Olhando a variável R² ajustada, ela apresentou uma melhora mínima, enquanto o AIC diminuiu minimamente, ao meu ver o modelo ficou melhor dado que eliminou-se uma variável que não apresenta tanto peso para o modelo.

````python
# Removendo tipo_residencia do modelo

y, X = patsy.dmatrices('np.log(renda) ~ C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda, Treatment("Assalariado")) + \
                       C(educacao, Treatment("Secundário")) + C(estado_civil, Treatment("Casado")) + idade + tempo_emprego + qt_pessoas_residencia + 0', data=df_1)

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

### Insights:

A variável ```educacao``` apresenta quatro categorias, e apenas uma delas apresenta significância estatística para o modelo, dessa forma ela também será removida do modelo.

````python
# Removendo educacao do modelo

y, X = patsy.dmatrices('np.log(renda) ~ C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda, Treatment("Assalariado")) + \
                       C(estado_civil, Treatment("Casado")) + idade + tempo_emprego + qt_pessoas_residencia + 0', data=df_1)

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

### Insights:

Remover ```educacao``` fez com que o R² e o R² ajustado tivessem valores menores, mas AIC maior.

As variáveis restantes apresentam mais categorias com significância estatística do que sem significância.

````python
# Removendo as variáveis que continuam apresentam p-value maior que 5%

# Removendo tipo_renda = Pensionista e Bolsista

df_2 = df_1[df_1['tipo_renda'] != 'Pensionista']
df_2 = df_2[df_2['tipo_renda'] != 'Bolsista']

# Removendo estado_civil = União

df_2 = df_2[df_2['estado_civil'] != 'União']
````

````python
y, X = patsy.dmatrices('np.log(renda) ~ C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda, Treatment("Assalariado")) + \
                       C(estado_civil, Treatment("Casado")) + idade + tempo_emprego + qt_pessoas_residencia + 0', data=df_2)

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

### Insights:

Removendo as categorias que apresentaram p-value acima de 5% observou-se uma boa melhora no R² e no R² ajustado, por outro lado o valor de AIC diminui, comparado com o último modelo.

Comparando o último modelo obtido com o primeiro, observa-se uma melhora tanto do R² e do R² ajustado, contudo o valor de AIC ficou menor que o inicial.
