### 1. Baixando dados de ativos

```python
# Importando bibliotecas

from pandas_datareader import data as web

import plotly.express as px
from plotly import graph_objects

import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
```

```python
# Carregando várias bases de dados usando get_data_yahoo
%%time

ativos = ['BBDC3.SA', 'MGLU3.SA', 'CMIG3.SA']
dt_ini = '2022-01-01'

# Se não colocar a data de fim, ele irá entender que a data de fim é a data mais recente

yf = web.get_data_yahoo(ativos, start=dt_ini)
yf.head()
```

### 2. Observado a volatilidade dentro do dia

1. Calcule a amplitude de variação do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo)
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Criando coluna de amplitude das ações

df = yf.stack().reset_index()
df['Amplitude'] = df['High'] - df['Low']
df.head()
```

```python
# Selecionando apenas as colunas que serão utilizadas no gráfico

colunas = ['Date', 'Ticker', 'Amplitude']

# Criando um dataframe compacto
df_01 = df[colunas]
df_01.head()
```

```python
# Plotando o gráfico

fig = px.line(df_01, x='Date', y='Amplitude', color='Ticker', template='simple_white')
fig.show()
```

### Insights:

- Observando o gráfico observa-se que as ações da CEMIG apresentaram maior volatilidade diária durante o período analisado sua maior variação foi de quase 2, no dia 22 de novembro de 2023.

- A maior variação diária observada na ação do Magazine Luisa foi de 0,78, e do Bradesco de 1,5.
