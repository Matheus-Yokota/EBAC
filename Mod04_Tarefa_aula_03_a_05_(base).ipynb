## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1.

# Para saber o percentual de acerto por aluno, por disciplina basta dividir os dataframes das matérias pela quantidade total de questões por disciplina.

porcentagem_matematica = df_mat / 24
porcentagem_portugues = df_por / 18
porcentagem_geografia = df_geo / 8
porcentagem_ingles = df_ing / 8
porcentagem_historia = df_his / 8
porcentagem_fisica = df_fis / 12
porcentagem_quimica = df_qui / 12

# Como exemplo do output usarei a matéria de matemática.

porcentagem_matematica.round(decimals= 2) # Utilizei o método round para arredondar o valor para duas casas decimais.
````

````python
# 2.

# Para saber o total de acerto por aluno basta somar todos os dataframes.

total_alunos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui

total_alunos
````

````python
# 3.

# Para saber a porcentagem geral de cada aluno basta dividir a quantidade de acerto de cada aluno pela quantidade total de questões

total_questões = 24 + 18 + 8 + 8 + 8 + 12 + 12

porcentagem_geral = total_alunos / total_questões

porcentagem_geral.round(decimals= 2)
````

````python
# 4.

# O enunciado pede a quantidade de alunos que acertaram MAIS de 45 questões.

# Peguei o dataframe da quantidade de acertos por aluno, filtrei para aqueles que acertaram mais que 45 questões, e realizei a contagem.
total_alunos[total_alunos['Qt_acertos'] > 45].value_counts()

# Aqui é apresentado a quantidade de alunos que acertou mais que 45 questões por quantidade de questões acertadas, se quisermos saber o total basta colocar um .sum() no final
````

````python
total_alunos[total_alunos['Qt_acertos'] > 45].value_counts().sum()
````

````python
# Como eu estava com dificuldade de realizar essa segunda parte achei mais fácil colocar todos os dados em um único dataframe.


df_total = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui # Soma o total de questões acertadas para cada aluno.

df_geral = df_mat
df_geral['Qt_acertos_mat'] = df_mat['Qt_acertos']
df_geral['Qt_acertos_por'] = df_por['Qt_acertos']
df_geral['Qt_acertos_geo'] = df_geo['Qt_acertos']
df_geral['Qt_acertos_his'] = df_his['Qt_acertos']
df_geral['Qt_acertos_ing'] = df_ing['Qt_acertos']
df_geral['Qt_acertos_fis'] = df_fis['Qt_acertos']
df_geral['Qt_acertos_qui'] = df_qui['Qt_acertos']
df_geral['Total_acertos'] = df_total['Qt_acertos']

df_geral.drop(columns='Qt_acertos', inplace=True)
df_geral
````

````python
# 1.

# Para saber quais alunos foram reprovados, filtrei o dataframe.

df_reprovados = df_geral[(df_geral['Qt_acertos_mat'] == 0) |
                         (df_geral['Qt_acertos_fis'] == 0) |
                         (df_geral['Qt_acertos_qui'] == 0)]

df_reprovados
````

````python
# 2.

# Para saber a quantidade de alunos que reprovaram utilizei o método shape que dá o formato do dataframe.

df_reprovados.shape
````

````python
# 3.

# Filtrei as notas de história dos alunos reprovados e utilizei o método mean() para obter a média.

df_reprovados_his = df_reprovados['Qt_acertos_his']

df_reprovados_his.mean()
````

````python
# 3.

# Filtrei as notas de geografia dos alunos reprovados e utilizei o método mean() para obter a média.

df_reprovados_geo = df_reprovados['Qt_acertos_geo']

df_reprovados_geo.mean()
````

````python
# 4.

# Para saber quais alunos não reprovaram utilizei o método drop na base de dados.
# Depois filtrei pela nota de corte, > 45 acertos.

reprovados_index = df_reprovados.index

df_nao_reprovados = df_geral.drop(reprovados_index)

df_aprovados = df_nao_reprovados[df_nao_reprovados['Total_acertos'] > 45]

df_aprovados
````
