```python
# Aqui eu importo as bibliotecas que serão utilizadas.

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# %matplotlib notebook
```

```python
# Coloco o caminho do csv em uma variável.

arquivo = r"C:\Users\kenji\OneDrive\Documentos\EBAC\Ciêntista de dados\Módulo 6\3f77c7b9de\SINASC_RO_2019.csv"
```

```python
# Carrego o meu dataframe.

df = pd.read_csv(arquivo, sep=',')
```

```python
# Visualizo quais são as minhas variáveis com o info()

df.info()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Usando o seaborn realizo o plot do boxplot para a idade da mãe.
# Dá para ver que nesta base de dados existem mães com idades bem elevadas, mais de 50 anos.
# Outra coisa que dá para notar é que existem mães bem jovens, perto de 10 anos de idade.

sns.boxplot(y='IDADEMAE', data=df)
```

####  2) Faça um boxplot do peso do bebê

```python
# Aqui realizo o plot do boxplot para o peso com que os bebês nascem.
# O peso é medido em gramas.
# Existem muitos dados fora boxplot.

sns.boxplot(y='PESO', data=df)
```

####  3) Faça um histograma da idade da mãe

```python
# Aqui realizo o plot do histograma da idade das mães.
# O histograma está deslocado mais para esquerda, ou seja, as mães presentes nesta base de dados são jovens.

sns.histplot(x='IDADEMAE', data=df, bins=20, element='step', kde=True)
```

#### 4) Faça um histograma do peso do bebê

```python
# Aqui realizo o plot do histograma do peso dos bebês após o nascimento.
# Difente do que foi observado no histograma da idade mães, o histograma do peso dos bebês está mais simétrica, evidenciando uma distribuição mais uniforme desses dados.

sns.histplot(x='PESO', data=df, bins=20, element='step', kde=True)
```
